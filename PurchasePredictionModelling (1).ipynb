```python
import pandas as pd

customer_df = pd.read_csv('customer_purchase_data.csv')
```

```python
customer_df
```

```python
categories = ['Electronics', 'Clothing', 'Home Goods', 'Beauty', 'Sports']

customer_df['ProductCategory'] = customer_df['ProductCategory'].apply(lambda x: categories[x])

customer_df = pd.get_dummies(customer_df, columns = ['ProductCategory'], dtype='int')
```

```python
customer_df
```

```python
customer_df.describe()
```

```python
customer_df.corr()
```

```python
customer_df = customer_df.drop(['Gender'], axis=1)
```

```python
X, y = customer_df[[
    'Age',
    'AnnualIncome',
    'NumberOfPurchases',
    'TimeSpentOnWebsite',
    'LoyaltyProgram',
    'DiscountsAvailed',
    'ProductCategory_Beauty',
    'ProductCategory_Clothing',
    'ProductCategory_Electronics',
    'ProductCategory_Home Goods',
    'ProductCategory_Sports',
]].values, customer_df['PurchaseStatus'].values
X_cols = customer_df[[
    'Age',
    'AnnualIncome',
    'NumberOfPurchases',
    'TimeSpentOnWebsite',
    'LoyaltyProgram',
    'DiscountsAvailed',
    'ProductCategory_Beauty',
    'ProductCategory_Clothing',
    'ProductCategory_Electronics',
    'ProductCategory_Home Goods',
    'ProductCategory_Sports',
]]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)
```

```python
import xgboost as xgb
```

```python
from hyperopt import STATUS_OK, Trials, fmin, hp, tpe

space={'max_depth': hp.quniform("max_depth", 3, 18, 1),
        'gamma': hp.uniform ('gamma', 0,9),
        'reg_alpha' : hp.quniform('reg_alpha', 0,180,1),
        'reg_lambda' : hp.uniform('reg_lambda', 0,1),
        'colsample_bytree' : hp.uniform('colsample_bytree', 0.5,1),
        'min_child_weight' : hp.quniform('min_child_weight', 0, 10, 1),
        'n_estimators': 180,
        'seed': 0
      }
```

```python
from sklearn.metrics import f1_score

def objective(space):
    clf=xgb.XGBClassifier(eval_metric="auc",early_stopping_rounds=10,
                    n_estimators =space['n_estimators'], max_depth = int(space['max_depth']), gamma = space['gamma'],
                    reg_alpha = int(space['reg_alpha']),min_child_weight=int(space['min_child_weight']),
                    colsample_bytree=int(space['colsample_bytree']))

    evaluation = [( X_train, y_train)]

    clf.fit(X_train, y_train,
            eval_set=evaluation,
            verbose=False)


    pred = clf.predict(X_test)
    ave_f1_score = f1_score(y_test, pred, average='weighted')
    return {'loss': -ave_f1_score, 'status': STATUS_OK }
```

```python
from hyperopt.early_stop import no_progress_loss

trials = Trials()

best_hyperparams = fmin(fn = objective,
                        space = space,
                        algo = tpe.suggest,
                        max_evals = 200,
                        trials = trials,
                        early_stop_fn=no_progress_loss(20))
```

```python
best_hyperparams
```

```python
xgb_classifier = xgb.XGBClassifier(
colsample_bytree= best_hyperparams['colsample_bytree'],
gamma=best_hyperparams['gamma'],
max_depth=int(best_hyperparams['max_depth']),
min_child_weight=best_hyperparams['min_child_weight'],
reg_alpha=best_hyperparams['reg_alpha'],
reg_lambda=best_hyperparams['reg_lambda']
)
xgb_classifier.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
predictions = xgb_classifier.predict(X_test)
print("Weighted Average F1 Score of the Model:",f1_score(y_test,predictions,average="weighted"))
```

```python
import shap

explainer = shap.Explainer(xgb_classifier)
shap_values = explainer(X_cols)

shap.plots.bar(shap_values, clustering_cutoff=1)
```
